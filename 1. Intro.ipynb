```python
# List
l = [4, 6, 2, 1]
# 0 1 2 3
# print(l[2])
# print(len(l)) # length
len(l)

# index property
l[3] # get order dari sebelah kiri

l[-3] # get order dari sebelah kanan
l[1:3] # get order from 1 until before 3
l[2:] # get order list from 2 until length

l[:3] # 

# mutable (manipulasi)
l.append(9) # menambahkan isi di belakang
l[-2] = 5 # menggannti value di n index
l[0] = 0

del l[2]

l.sort() # sort
# comment
# help(list.extend) # cari tau deskripsi dari suatu built in methods dari type data

l.extend([2, 3])

max(l) # special methods buat list atau tipe data lain
min(l)

# l.append([1, 2]) # nested
# ll = [[1, 2], [2, 2]]

# packing unpacking
[nol, lima, enam] = [0, 5, 6]
enam

```
