```python
# Generate data of people mass in pounds
# generate 2 distribusi normal

np.random.seed(42)

normal_distribution_us_male_mass_pounds = np.random.normal(loc=181, scale=24, size=6000)
normal_distribution_us_female_mass_pounds = np.random.normal(loc=132, scale=22, size=6500) # loc = mean, scale = std

all_mass_values = np.concatenate((normal_distribution_us_male_mass_pounds, normal_distribution_us_female_mass_pounds), axis=0)
df_ppl_mass = pd.DataFrame(data={'us_people_mass_pounds': all_mass_values })
df_ppl_mass
sns.distplot(df_ppl_mass['us_people_mass_pounds'], color='darkslategrey')
plt.xlabel('mass [pounds]', labelpad=14)
plt.ylabel('probability of occurence', labelpad=14)
plt.title('Distribution of Mass of People in US', y=1.015, fontsize=20)
```

```python
df_ppl_mass['us_people_mass_pounds'].mean()

std_dev_mass = df_ppl_mass['us_people_mass_pounds'].std()
std_dev_mass
```

```python
sample_means = []
n = 25
for sample in range(0, 300):
    # random sampling
    sample_values = np.random.choice(a=df_ppl_mass['us_people_mass_pounds'], size=n)
    sample_mean = np.mean(sample_values)
    sample_means.append(sample_mean)

# sampling distribution
sns.distplot(sample_means)
plt.title('Distribution of sample means ($n=25$) of Peoples mass in pounds', y=1.015, fontsize=20)
plt.xlabel('sample mean mass in pounds', labelpad=14)
plt.ylabel('Frequency of occurence', labelpad=14)
```

```python

import pandas as pd
import numpy as np
import statsmodels.api as sm
import scipy.stats.distributions as dist
df = pd.read_csv('Heart.csv')
df.head()
```

```python
# Step 1: Define hypothesis null and alternative
# H0: Pir = 0.42
# H1: Pus > 0.42

# step 2: Calculate the population proportion
proportion_us = len(df[df['target'] == 1])/len(df)
proportion_us
```

```python
# step 3 Calculate test statistic

# Best Estimate - Hypothesized Estimate / SE of Estimate
se = np.sqrt(0.42 * (1-0.42) / len(df))

be = proportion_us
he = 0.42
test_stat = (be-he) / se
test_stat
```

```python
# step 4: calculate the p-value
p_value = 2 * dist.norm.cdf(-np.abs(test_stat))
p_value

# p_value < alpha (0.05)
# conclusion: reject the null hypothesis where proportion_us > proportion_ireland
# us having more 42% from suffering heart disease
```
