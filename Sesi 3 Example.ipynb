```python
def change_me(dictionary, new_key, new_value):
    dictionary[new_key] = new_value

d = { 'first_name': 'michael', 'last_name': 'jackson' }
change_me(d, 'title', 'king of pop')
```

```python
def front_input(input_list, value):
    if type(value) == list:
        for i in range(len(value)):
            input_list.insert(0, value[i])
    else:
        input_list.insert(value)
    return

l = [5, 6, 7, 8]
front_input(l, [1,2,3,4])
print(l)
# print(type('123') == str)
# print(type(9) == int)
```

```python
# The anonymous function or Lambda function
# function definition is here
sum_lambda = lambda arg1, arg2: arg1 + arg2;

def sum_biasa(arg1, arg2):
    arg1 + arg2

print(sum_lambda(20, 20))
print(sum_biasa(30, 10))
```

```python
# Global vs. Local variables
money = 0

def dollar_to_rupiah(dollar):
    money = dollar * 14372
    print('Inside function: ', money)
    return money

dollar_to_rupiah(10)

print('Outside function: ', money)
```
