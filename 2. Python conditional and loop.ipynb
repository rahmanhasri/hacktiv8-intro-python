```
# looping

n = 5
while n > 0: # 5 print n, 4 > 0, 0 > 0
    print(n)
    # n = n - 1
    n -= 1 # cek lagi dikondisi while

while n < 1:
    n -= 1
    print(n)
    if n == -2:
        print('selesai')
        break

```

```
# continue
a = 0
while a < 5:
    a += 1
    if a == 2:
        continue # next looping
    print(a)
print('loop ended')


while a > 2:
    a -= 1 # a = a - 1
    print(a)
    if a == 3:
        break
else:
    print('loop done')

while True:
    print('infinite loop')

```
